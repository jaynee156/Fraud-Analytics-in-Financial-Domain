Building Stockflow graph


```
#building the Stockflow graph - a directed weighted Graph
Graph = nx.from_pandas_edgelist(df,source='Seller Id',target='Buyer Id', edge_attr='Amount', create_using=nx.DiGraph())
```

Finding KNN of every node

```
#out_edges_n is a dictionary of node: list of it's out edges with weights
out_edges_n={}
for node in Graph.nodes():
  out_edges_n[node] = list(Graph.out_edges(nbunch=node, data='Amount'))
```

```
len(out_edges_n)
```

```
# To sort the list of out_edges of every node
  
# get the last key. 
def last(n): 
    return n[m]   
   
# function to sort the tuple    
def sort(tuples):  
    return sorted(tuples, key = last) 
   
# to sort every node's list of out edges by edge weight 
m=2
for x in out_edges_n:
  out_edges_n[x] = sort(out_edges_n[x])
```

```
out_edges_n
```

```
k=4
knn_n = {}
for x in out_edges_n:
  knn_n[x] = set()
  length=len(out_edges_n[x])
  if length>k:
    for i in range(k):
      knn_n[x].add(out_edges_n[x][i][1])
  else:
     for j in range(length):
       knn_n[x].add(out_edges_n[x][j][1])
```

For every node u in the Graph, to remove every edge from u to v where v is not in knn(u)

```
all_edges=set()
knn_edges=set()
remove_edges=set()
ebunch=[]
for node in Graph.nodes():
  for p in out_edges_n[node]:
    all_edges.add((p[0],p[1]))
    if (knn_n[node] is not set()):
      for val in knn_n[node]:
        knn_edges.add((node,val))
```

```
#for all nodes: to removes edges which are from the given node to a node not in knn of the given node
remove_edges = all_edges - knn_edges
ebunch=list(remove_edges)
Graph.remove_edges_from(ebunch=ebunch)
```

So now every node in the graph has edges only to nodes that are in it's knn set


```
#in_edges_n is a dictionary of node: list of it's in edges with weights
in_edges_n={}
for node in Graph.nodes():
  in_edges_n[node] = list(Graph.in_edges(nbunch=node, data='Amount'))
```

```
#c is a cluster- a set of nodes
#function to find collusion index of a cluster
def co_index(c):
  if len(c) ==1 :
    return 0
  else:
    ic=0
    ec=0
    g=list(c)
    sg=Graph.subgraph(g)
    for u, v in sg.edges():
      ic += sg[u][v]['Amount']
    #adding weights of outedges from c to outside i.e u in cluster and v not in cluster
    for x in c:
      for y in out_edges_n[x]:
        if (y[1] not in c):
          ec += y[2]
    # adding weights for in edges into the cluster, i.e u not in cluster but v in cluster
      for y in in_edges_n[x]:
        if (y[0] not in c):
          ec += y[2]
    if (ec == 0):
      phi=ic/1
    else:
      phi=ic/ec
    return phi
```

```
#function to find collusion level of two clusters
def LC(c1, c2):
  return co_index(c1.union(c2))
```

```
#function to sort pairs of clusters by their LC val in descending order
def Sort_Tuple(tup):  
    return (sorted(tup, key = lambda x: x[1], reverse=True)) 
```

```
#kmh compatibility of two clusters (each cluster is basically a set of nodes)
def kmh(s1, s2):
  k=4
  m=1
  h=0.70
  #count of nodes in cluster s1 that are km compatible with cluster s2
  km_s1=0
  #count of nodes in cluster s2 that are km compatible with cluster s1
  km_s2=0
  for x in s1:
    count=0
    for p in knn_n[x]:
      if(p in s2):
        count+=1
    if count >= m :
      km_s1 += 1
  for x in s2:
    count =0
    for p in knn_n[x]:
      if(p in s1):
        count +=1
    if count>=m :
      km_s2 +=1
  if( round(km_s1/len(s1),2)>= h and round(km_s2/len(s2),2)>=h):
    return 1
  else:
    return 0
  
```

Main part of the collusion clustering algorithm

```
#S is the list of collusion sets
S =[]
for node in Graph.nodes():
  S.append({node})
```

```
len(S)
```

```

B=[]
while(1):
  flag=0

  for i in range(0,len(S)-1):
    for j in range(i+1,len(S)):
      B.append(((S[i],S[j]),LC(S[i],S[j])))
      B=Sort_Tuple(B)

  for x in B:
    if (x[1]>0 and kmh(x[0][0],x[0][1])):
      flag=1
      S.remove(x[0][0])
      S.remove(x[0][1])
      S.append(x[0][0].union(x[0][1]))
      break
  B.clear()


  
```
